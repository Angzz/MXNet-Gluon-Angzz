```python
%matplotlib inline
from mxnet import nd
from mxnet import image
import matplotlib.pyplot as plt

rgb_mean = nd.array([123, 117, 104])


def box_to_rec(box, color, linewidth=2):
    rec = plt.Rectangle((box['xmin'], box['ymin']), width=box['xmax']-box['xmin'], height=box['ymax']-box['ymin'],
                       fill=False, edgecolor=color, linewidth=linewidth)
    return rec

bb = {'xmin':34.535904/512*500, 'ymin':0.000000/512*375, 'xmax':475.524292/512*500, 'ymax':405.266510/512*375}

# 显示图片
def show_images(img, data_shape, figsize=None):
    
    plt.figure(figsize=figsize)
    plt.imshow(img.asnumpy())

    # 画边框
    rec = box_to_rec(bb, 'red', 2)
    plt.gca().add_patch(rec)
```

```python
img = image.imread('2008_000002.jpg', 1)
print(img.shape)
show_images(img, data_shape=1024, figsize=(6,6))
```

```python
img_flip = nd.flip(img, 1)
show_images(img_flip, data_shape=1024, figsize=(6,6))
```
